```python
num=int(input("Enter Number of processes ==> ")) #number of processes
service_time = [0] * num  
service_time[0] = 0

#burst time
burst=[]
for i in range(0,num):
    print("Enter burst time for Process",i,":")
    bt=int(input())
    burst.append(bt)

#arrival time
arrival=[]
for i in range(0,num):
    print("Enter arrival time for Process",i+1,":")
    at=int(input())
    arrival.append(at)
#waiting time    
wt=[0]
for i in range(1, num):  
          
        # Add burst time of previous processes  
        service_time[i] = (service_time[i - 1] + burst[i - 1])  
  
        # Find waiting time for current 
        # process = sum - at[i]  
        wt.append(service_time[i] - arrival[i])  
  
        # If waiting time for a process is in  
        # negative that means it is already  
        # in the ready queue before CPU becomes  
        # idle so its waiting time is 0  
        if (wt[i] < 0): 
            wt[i] = 0  
print("waiting time is:",wt)   

# turnaround time 
tat=[]    
for i in range(num):
    tat.append(wt[i]+burst[i])
print("\nturn around time is :",tat) 


#total tat 
total_tat=0
for i in range(num):
    total_tat=tat[i]+total_tat
print("\naverage turn around time is:",total_tat/num)        

#total waiting time
total_wt=0
for i in range(num):
    total_wt=total_wt+wt[i]
print("\naverage waiting time is:%.5f "%(total_wt /num))
```
